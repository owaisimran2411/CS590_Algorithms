# **Naive Insertion Sort Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your own data)
data_array = np.array([10000, 25000, 50000, 100000, 250000, 1000000])
X  = (data_array - data_array.min()) / (data_array.max() - data_array.min())
A = np.array([1.3553, 9.0436, 35.6904, 229.4818, 1850.282, 5184.3125])
B = np.array([.0009, .0022, .0028, .0076, .019, .056])
C = np.array([2.8972,  16.6688,  85.1778,  338.2118,  2274.1594,  12234.9776])

# Create a scatter plot
plt.figure(figsize=(5, 6))

plt.scatter(X, A, label='Random Vector', color='r', marker='o', s=50)
plt.scatter(X, B, label='Sorted Vector', color='g', marker='s', s=50)
plt.scatter(X, C, label='Inverse Sorted Vector', color='b', marker='^', s=50)

plt.plot(X, A, linestyle='--', color='r')
plt.plot(X, B, linestyle='--', color='g')
plt.plot(X, C, linestyle='--', color='b')

plt.xlabel('M (Normalized)')
plt.ylabel('T(n)')
plt.title('Algorithm: Naive Insertion Sort for n=10')
plt.legend()

correlation_matrix_A = np.corrcoef(X, A)
correlation_coefficient_A = correlation_matrix_A[0, 1]
r_squared_A = correlation_coefficient_A**2
plt.text(0.1, 0.7, f'R = {correlation_coefficient_A:.2f}', fontsize=12, color='r', transform=plt.gca().transAxes)
plt.text(0.1, 0.65, f'R^2 = {r_squared_A:.2f}', fontsize=12, color='r', transform=plt.gca().transAxes)

correlation_matrix_B = np.corrcoef(X, B)
correlation_coefficient_B = correlation_matrix_B[0, 1]
r_squared_B = correlation_coefficient_B**2
plt.text(0.1, 0.6, f'R = {correlation_coefficient_B:.2f}', fontsize=12, color='g', transform=plt.gca().transAxes)
plt.text(0.1, 0.55, f'R^2 = {r_squared_B:.2f}', fontsize=12, color='g', transform=plt.gca().transAxes)

correlation_matrix_C = np.corrcoef(X, C)
correlation_coefficient_C = correlation_matrix_C[0, 1]
r_squared_C = correlation_coefficient_C**2
plt.text(0.1, 0.5, f'R = {correlation_coefficient_C:.2f}', fontsize=12, color='b', transform=plt.gca().transAxes)
plt.text(0.1, 0.45, f'R^2 = {r_squared_C:.2f}', fontsize=12, color='b', transform=plt.gca().transAxes)

plt.grid(True)


plt.show()
```
